#number of cars per capita

```python
cars_df = pd.read_csv('https://raw.githubusercontent.com/MerienMoelker/Final-assignment---CO2-emissions/main/new-vehicles-type-area%20(1).csv')
pop_df = pd.read_csv('https://raw.githubusercontent.com/MerienMoelker/Final-assignment---CO2-emissions/main/population-since-1800.csv')
cars_df = cars_df.merge(pop_df, left_on=['Entity', 'Code', 'Year'], right_on=['Entity', 'Code', 'Year'], how='outer')
cars_df['total_new_cars'] = cars_df['battery_electric_number'] + cars_df['plugin_hybrid_number'] + cars_df['full_mild_hybrid_number'] + cars_df['petrol_number'] + cars_df['diesel_gas_number']
cars_df.loc[:,'battery_electric_number'] = cars_df.loc[:,'battery_electric_number'] / cars_df['Population (historical estimates)']
cars_df.loc[:,'plugin_hybrid_number'] = cars_df.loc[:,'plugin_hybrid_number'] / cars_df['Population (historical estimates)']
cars_df.loc[:,'full_mild_hybrid_number'] = cars_df.loc[:,'full_mild_hybrid_number'] / cars_df['Population (historical estimates)']
cars_df.loc[:,'petrol_number'] = cars_df.loc[:,'petrol_number'] / cars_df['Population (historical estimates)']
cars_df.loc[:,'diesel_gas_number'] = cars_df.loc[:,'diesel_gas_number'] / cars_df['Population (historical estimates)']
cars_df.loc[:,'total_new_cars'] = cars_df.loc[:,'total_new_cars'] / cars_df['Population (historical estimates)']
cars_df.drop(columns=['Population (historical estimates)'], inplace=True)
df_num_of_cars = co2_df.merge(cars_df, left_on=['Entity', 'Code', 'Year'], right_on=['Entity', 'Code', 'Year'], how='outer')
```
