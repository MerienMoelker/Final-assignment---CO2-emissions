```python
import pandas as pd
import numpy as mp
import matplotlib as mlt
import matplotlib.pyplot as plt
```

```python
co2_df = pd.read_csv('https://raw.githubusercontent.com/MerienMoelker/Final-assignment---CO2-emissions/main/annual-co2-emissions-per-country.csv')
pop_df = pd.read_csv('https://raw.githubusercontent.com/MerienMoelker/Final-assignment---CO2-emissions/main/population-since-1800.csv')
df = co2_df.merge(pop_df, left_on=['Entity', 'Code', 'Year'], right_on=['Entity', 'Code', 'Year'], how='outer')
df['co2_per_capita'] = df['Annual CO₂ emissions'] / df['Population (historical estimates)']
```

```python
df.dropna(inplace=True)
df
```

```python
df_grp = df.groupby(by=['Entity'])
```

```python
new_df = df_grp['co2_per_capita'].agg(['max', 'last'])
```

```python
new_df['change'] = 100 - (new_df['last'] / new_df['max']) * 100
```

```python
filter = new_df['last'] > 2
display(new_df[filter].shape)
new_df[filter].sort_values(by='change', ascending=False).reset_index().loc[0:9,['Entity', 'change']].head(10)
```
